```python
grouped_sum = df.groupby('Sub-Category')[['Profit', 'Sales']].sum().reset_index()
grouped_mean = df.groupby('Sub-Category')[['Profit', 'Sales']].mean().reset_index()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))


sbn.barplot(data=grouped_sum.melt(id_vars='Sub-Category', var_name='Metric', value_name='Value'),  
            x='Sub-Category', y='Value', hue='Metric', ax=axes[0])
axes[0].set_title('Total Profit and Sales per Sub-Category')
axes[0].set_ylabel('Total Value')
axes[0].tick_params(axis='x', rotation=90)


sbn.barplot(data=grouped_mean.melt(id_vars='Sub-Category', var_name='Metric', value_name='Value'),  
            x='Sub-Category', y='Value', hue='Metric', ax=axes[1])
axes[1].set_title('Mean Profit and Sales per Sub-Category')
axes[1].set_ylabel('Mean Value')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
Staples = df[df['Product Name'] == 'Staples']
Staples = Staples.set_index('Order Date')  
Staples = Staples['Sales'].resample('M').sum()
Staples = Staples.reset_index()

Avery_Non_Stick_Binders = df[df["Product Name"] == "Avery Non-Stick Binders"]
Avery_Non_Stick_Binders = Avery_Non_Stick_Binders.set_index('Order Date')
Avery_Non_Stick_Binders = Avery_Non_Stick_Binders['Sales'].resample('M').sum()
Avery_Non_Stick_Binders = Avery_Non_Stick_Binders.reset_index()

KI_Adjustable_Height_Table = df[df["Product Name"] == "KI Adjustable-Height Table"]
KI_Adjustable_Height_Table = KI_Adjustable_Height_Table.set_index('Order Date')
KI_Adjustable_Height_Table = KI_Adjustable_Height_Table['Sales'].resample('M').sum()
KI_Adjustable_Height_Table = KI_Adjustable_Height_Table.reset_index()

Storex_Dura_Pro_Binders = df[df["Product Name"] == "Storex Dura Pro Binders"]
Storex_Dura_Pro_Binders = Storex_Dura_Pro_Binders.set_index('Order Date')
Storex_Dura_Pro_Binders = Storex_Dura_Pro_Binders['Sales'].resample('M').sum()
Storex_Dura_Pro_Binders = Storex_Dura_Pro_Binders.reset_index()

plt.figure(figsize=(20, 13))
plt.plot(Staples['Order Date'], Staples['Sales'], marker='o', label='Staples')
plt.plot(Avery_Non_Stick_Binders['Order Date'], Avery_Non_Stick_Binders['Sales'], marker='o', label='Avery Non-Stick')
plt.plot(KI_Adjustable_Height_Table['Order Date'], KI_Adjustable_Height_Table['Sales'], marker='o', label='KI Table')
plt.plot(Storex_Dura_Pro_Binders['Order Date'], Storex_Dura_Pro_Binders['Sales'], marker='o', label='Dura Pro Binders')

plt.title('Products Monthly Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
