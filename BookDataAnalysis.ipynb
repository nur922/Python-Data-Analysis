# Author: The name of the author of the book. (Text)
language_code: The code representing the language in which the book is written. (Text)
Author_Rating: The rating of the author based on their previous works. (Numeric)
Book_average_rating: The average rating given to the book by readers. (Numeric)
Book_ratings_count: The number of ratings given to the book by readers. (Numeric)
genre: The genre or category to which the book belongs. (Text)
gross sales: The total sales revenue generated by a specific book. (Numeric)
publisher revenue: The revenue earned by a publisher from selling a specific book. (Numeric)
sale price: The price at which a specific book was sold. (Numeric)
sales rank: The rank of a particular book based on its sale performance. (Numeric)
units sold: The number of units sold for any particular book. (Numeric)

```python
publisher_unit_sold=df.groupby("Publisher")["units sold"].sum()
plt.figure(figsize=(13,9))
ax =sbn.barplot(x=publisher_unit_sold.values , y=publisher_unit_sold.index , data=df ,palette="pastel" )
plt.title("Units Sold by Publisher")
plt.xlabel("Amount")
plt.ylabel("Publisher")
ax.bar_label(ax.containers[0], fmt='%d')
plt.xticks(rotation=90)    
plt.tight_layout()
plt.show()
```

```python
publisher_gross_sold=df.groupby("Publisher")["gross sales"].sum()
plt.figure(figsize=(13,9))
ax =sbn.barplot(x=publisher_gross_sold.values , y=publisher_gross_sold.index , data=df ,palette="pastel" )
plt.title("Gross Sold by Publisher")
plt.xlabel("Amount")
plt.ylabel("Publisher")
ax.bar_label(ax.containers[0], fmt='%d')
plt.xticks(rotation=90)    
plt.tight_layout()
plt.show()
```

```python
total_gross_sales_by_author = df.groupby('Author')['gross sales'].sum().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(15, 8))

bars = total_gross_sales_by_author.head(20).plot(kind='bar', color='pink', ax=ax)

ax.set_xlabel('Author')
ax.set_ylabel('Total Gross Sales')
ax.set_title('Total Gross Sales For Each Author')

ax.bar_label(ax.containers[0], fmt='%d')

plt.show()
```

```python


plt.figure(figsize=(13, 9))

units_per_year = df.groupby('Publishing Year')['units sold'].sum()

ax = units_per_year.plot(kind='line', marker='o', color="orange")

plt.xlabel('Publishing Year')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold Over the Years')

for x, y in zip(units_per_year.index, units_per_year.values):
    ax.annotate(f'{int(y)}', xy=(x, y), xytext=(0, 5), textcoords='offset points', ha='center', fontsize=9)

plt.show()
```

```python
plt.figure(figsize=(13, 9))

gross_per_year = df.groupby('Publishing Year')['gross sales'].sum()

ax = gross_per_year.plot(kind='line', marker='o', color="orange")

plt.xlabel('Publishing Year')
plt.ylabel('Total gross Sales')
plt.title('Total gross Sales Over the Years')

for x, y in zip(gross_per_year.index, gross_per_year.values):
    ax.annotate(f'{int(y)}', xy=(x, y), xytext=(0, 5), textcoords='offset points', ha='center', fontsize=9)

plt.show()
```

```python
book_gross_sold=df.groupby("Book Name")["gross sales"].sum().sort_values(ascending=False).head(20)
plt.figure(figsize=(13,9))
ax =sbn.barplot(x=book_gross_sold.values , y=book_gross_sold.index , data=df ,palette="pastel" )
plt.title("Most Sold out Books")
plt.xlabel("Amount")
plt.ylabel("Book")
ax.bar_label(ax.containers[0], fmt='%d')
plt.xticks(rotation=90)    
plt.tight_layout()
plt.show()
```
