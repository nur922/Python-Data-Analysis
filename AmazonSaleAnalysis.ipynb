# En Ã§ok satÄ±ÅŸ yapan kategoriler

```python
best_selling_category=df.groupby('Category')["Amount"].sum().sort_values(ascending=False).head()
print(best_selling_category)
```

```python
best_selling_category.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('En Ã‡ok SatÄ±ÅŸ Yapan 5 Kategori')
plt.xlabel('Kategori')
plt.ylabel('Toplam SatÄ±ÅŸ (Amount)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
best_category=df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
plt.figure(figsize=(15,8))
ax=sbn.barplot(x=best_category.values, y=best_category.index, data=df)
plt.title("Total Amount ($) by Categorys")
plt.xticks(rotation=45)
plt.grid(axis="y")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
top_5_cities = df['ship-city'].value_counts().nlargest(5).index
df_top_cities = df[df['ship-city'].isin(top_5_cities)]
plt.figure(figsize=(15,8))
ax = sbn.countplot(data=df_top_cities, x="Category", hue="ship-city", palette='tab10')  
plt.title("Category order by Top 5 Ship Cities")
plt.xticks(rotation=45)
plt.grid(axis="y")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
df["Size"].value_counts()
```

```python
top_5_category = df['Category'].value_counts().nlargest(5).index
df_top_category = df[df['Category'].isin(top_5_category)]
plt.figure(figsize=(15,8))
ax = sbn.countplot(data=df_top_category, x="Category", hue="Size", palette='tab10')  
plt.title("Category order by Size")
plt.xticks(rotation=45)
plt.grid(axis="y")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
plt.figure(figsize=(10,6))
sbn.histplot(data=df, x="Amount", hue="Category", bins=30, kde=True, palette="BrBG")
plt.title("Histogram of Amount by Category")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.show()
```

# Tarihe gÃ¶re en Ã§ok satÄ±lan Ã¼rÃ¼nler

```python
sales_Date = df.groupby('Date')['Amount'].sum()
print(sales_Date)
```

```python
df["month"].unique()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["month"]=df["Date"].dt.month_name()
plt.figure(figsize=(10,5))
month_order = [ 'March', 'April', 'May', 'June']
monthly_sales = df.groupby('month')['Amount'].sum().reindex(month_order, fill_value=0)
ax = monthly_sales.plot(kind='line', marker='o', color='#D2691E')
ax.set_title("Total Amount ($) by Months")
ax.set_xlabel("Month")
ax.set_ylabel("Total Amount ($)")
ax.set_xticks(range(len(month_order)))
ax.set_xticklabels(month_order, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
weekly_sales = df.groupby(pd.Grouper(key='Date', freq='W'))['Amount'].sum()
plt.figure(figsize=(12, 5))
weekly_sales.plot()
plt.title('HaftalÄ±k SatÄ±ÅŸ TutarÄ±')
plt.xlabel('Tarih')
plt.ylabel('Toplam SatÄ±ÅŸ (Amount)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
