```python
import numpy as np
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("AmazonSaleReport.csv")
```

```python
df=df.drop("fulfilled-by" , axis=1)
```

```python
df["Courier Status"].fillna(df["Courier Status"].mode()[0] , inplace=True)
```

```python
df=df.drop(["B2B","Unnamed: 22","index"], axis=1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["currency"].fillna(df["currency"].mode()[0], inplace=True)
```

```python
df["Amount"].fillna(df["Amount"].mode()[0], inplace=True)
```

```python
df['ship-city'].fillna(df['ship-city'].mode()[0],inplace=True)
df['ship-state'].fillna(df['ship-state'].mode()[0],inplace=True)
df['ship-postal-code'].fillna(df['ship-postal-code'].mode()[0],inplace=True)
df['ship-country'].fillna(df['ship-country'].mode()[0],inplace=True)
df['promotion-ids'].fillna(df['promotion-ids'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

# En Ã§ok satÄ±ÅŸ yapan kategoriler

```python
best_selling_category=df.groupby('Category')["Amount"].sum().sort_values(ascending=False).head()
print(best_selling_category)
```

```python
best_selling_category.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('En Ã‡ok SatÄ±ÅŸ Yapan 5 Kategori')
plt.xlabel('Kategori')
plt.ylabel('Toplam SatÄ±ÅŸ (Amount)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sbn.histplot(data=df, x="Amount", hue="Category", bins=30, kde=True, palette="BrBG")
plt.title("Histogram of Amount by Category")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.legend(title='Category')
plt.show()
```

# Tarihe gÃ¶re en Ã§ok satÄ±lan Ã¼rÃ¼nler

```python
sales_Date = df.groupby('Date')['Amount'].sum()
print(sales_Date)
```

```python
weekly_sales = df.groupby(pd.Grouper(key='Date', freq='W'))['Amount'].sum()
plt.figure(figsize=(12, 5))
weekly_sales.plot()
plt.title('HaftalÄ±k SatÄ±ÅŸ TutarÄ±')
plt.xlabel('Tarih')
plt.ylabel('Toplam SatÄ±ÅŸ (Amount)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Åžehirlere gÃ¶re en Ã§ok satÄ±lan Kategoriler

```python
city_sales=df.groupby("ship-city")["Amount"].sum().sort_values(ascending=False).head()
print(city_sales)
```

# Aylara gÃ¶re satÄ±ÅŸlar

```python
monthly_sales = df.groupby(pd.Grouper(key='Date', freq='M'))['Amount'].sum().reset_index()
plt.bar(monthly_sales['Date'], monthly_sales['Amount'], color='#73361C')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
```

# Durumlara gÃ¶re sipariÅŸ sayÄ±sÄ±

```python
status_counts = df['Status'].value_counts()

plt.figure(figsize=(12, 10))
status_counts.plot(kind='bar', color='#73361C')
plt.title('Order Count by Status')
plt.xlabel('Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# SatÄ±ÅŸlara gÃ¶re Åžehirler

```python
city_sales=df["ship-city"].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.pie(city_sales, 
        labels=city_sales.index, 
        autopct='%1.1f%%', 
        startangle=140
        )

plt.title('Top 10 Shipping Cities by Order Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sbn.kdeplot(df["Amount"], fill=True, color="#73361C")
plt.title("Density Plot of Amount")
plt.xlabel("Amount")
plt.ylabel("Density")
plt.show()
```
